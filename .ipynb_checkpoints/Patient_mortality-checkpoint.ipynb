## Outliers

Though we cannot clearly point out the outliers by just observing the histograms, however, we will use **z-score** method to appropriately retreive the number of outliers present in each of the variables. We will specify *threshold = 4* which is passed as an arbitrary value. This means that any data point whose z-score is greater than 4 standard deviations from the mean will be considered as an outlier. 


```python
df = cancer

# Calculate z-score for each variable
z_scores = np.abs((df - df.mean()) / df.std())

# Define threshold
threshold = 4

# Identify outliers for each variable
outliers = (z_scores > threshold).sum()

# Report number of outliers for each variable
print("Number of outliers for each variable: ")
print(outliers)
```

This indicates that majority of the variables are containing outliers except for a few of them such as *PctPrivateCoverage*, *PctPrivateCoverageAlone*, *PctEmpPrivCoverage*, *PctPublicCoverage*, etc. Also note, this might be the case since the variables follow normal distribution which is less prone to outliers.

# **Exercise 2**

Create an ML pipeline using scikit-learn (as demonstrated in the lab notebooks) to pre-process the training data. (3 marks)

# **Sample Answer to Exercise 2**

```python
# Creating a train and test split from the data.

from sklearn.model_selection import train_test_split

train_X, val_X, train_Y, val_Y = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
# Creating a pipeline

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy="median")),
    ('std_scaler',StandardScaler()),
])


# Transforming the data (imputing null values, scaling)
cancer_transformed = pipeline.fit_transform(cancer)
train_X_transformed = pipeline.fit_transform(train_X)
val_X_transformed = pipeline.fit_transform(val_X)

```

# **Sample Answer to Exercise 3**

We will use **GridSearchCV** from *scikit-learn* library to find the optimal regularization weights for Lasso and Ridge regression models. This function evalutates the model using cross validation to find the best parameter values. Using the identified weights we will train both the models along with linear regression (OLS), then evaluate and compare their performance using *RMSE* and *Cross validation*. 

**1. Lasso regression**

```python
# Using GridSearchCV to find the best parameters for lasso 

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso

param_grid_lasso = [{'alpha': [0.001, 0.01, 0.1, 1, 10, 100] }]

lasso_GSCV = Lasso()

# Here we set the cross validation k folds to 5
lasso_grid_search = GridSearchCV(lasso_GSCV, param_grid_lasso, cv=5)

# Training the model
lasso_grid_search.fit(train_X_transformed, train_Y)


```

```python
# The weight due to which the Lasso model generates better score
lasso_grid_search.best_params_
```

Scores for each of the parameters are available below:

**2. Ridge regression**

```python
# Using GridSearchCV to find the best parameters for ridge

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

param_grid_ridge = [{'alpha': [0.001, 0.01, 0.1, 1, 10, 100] }]

ridge_GSCV = Ridge()

# Here we set the cross validation n-folds to 5
ridge_grid_search = GridSearchCV(ridge_GSCV, param_grid_ridge, cv=5)

# Training the model
ridge_grid_search.fit(train_X_transformed, train_Y)
```

```python
# The weight due to which the Ridge model generates better score
ridge_grid_search.best_params_
```

## Linear Regression (Ordinary Least Squares)

```python
# Fitting a Linear Regression model (OLS: Ordinary Least Squares)
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

linear = LinearRegression()

# Training the model with train data
linear.fit(train_X_transformed, train_Y)

# Making predictions
linear_predictions = linear.predict(val_X_transformed)

# Evaluating the model using RMSE
linear_mse=mean_squared_error(val_Y, linear_predictions)
linear_rmse=np.sqrt(linear_mse)
print(linear_rmse) 
```

## Lasso Regression

```python
# Fitting Lasso model

lasso = Lasso(alpha=0.1)

# Training the model
lasso.fit(train_X_transformed, train_Y)

# Making predictions
lasso_predictions = lasso.predict(val_X_transformed)

# Evaluating the model using RMSE
lasso_mse=mean_squared_error(val_Y, lasso_predictions)
lasso_rmse=np.sqrt(lasso_mse)
print(lasso_rmse)
```

## Ridge Regression

```python
# Fitting Ridge model

ridge = Ridge(alpha=10)

# Training the model
ridge.fit(train_X_transformed, train_Y)

# Making predictions
ridge_predictions = ridge.predict(val_X_transformed)

# Evaluating the model using RMSE
ridge_mse=mean_squared_error(val_Y, ridge_predictions)
ridge_rmse=np.sqrt(ridge_mse)
print(ridge_rmse)
```

Hence, the weights using which both Lasso and Ridge regression models perform better have now been obtained. These weights have been identified using these steps:
- Using *GridSearchCV* technique from *scikit-learn* library.
- We define *n* number of parameters to be passed to the model.
- The model is evaluated using cross validation of 5 folds.
- The model is trained independently for every parameters that have been passed.
- Finally the performance of the model is evaluated for each of the parameters. The best parameter value is revealed for which the model has generated a better score.

The best value of 'alpha' for Lasso model is **0.1**.
The best value of 'alpha' for Ridge model is **10**.

## Cross Validation
Using cross validation of 5 folds, we will evaluate and compare the performance of all the three models defined above.

```python
# Linear Regression

from sklearn.model_selection import cross_val_score, cross_val_predict

linear_cv_scores = cross_val_score(linear, train_X_transformed, train_Y, scoring="neg_mean_squared_error",cv=5)

linear_rmse_scores = np.sqrt(-linear_cv_scores)

def display_scores(scores):
    print('Scores:',scores)
    print('Mean:',scores.mean())
    
display_scores(linear_rmse_scores)

```

```python
# Lasso Regression

from sklearn.model_selection import cross_val_score, cross_val_predict

lasso_cv_scores = cross_val_score(lasso, train_X_transformed, train_Y, scoring="neg_mean_squared_error",cv=5)

lasso_rmse_scores = np.sqrt(-lasso_cv_scores)

# Display mean score
display_scores(lasso_rmse_scores)
```

```python
# Ridge Regression

from sklearn.model_selection import cross_val_score, cross_val_predict

ridge_cv_scores = cross_val_score(ridge, train_X_transformed, train_Y, scoring="neg_mean_squared_error",cv=5)

ridge_rmse_scores = np.sqrt(-ridge_cv_scores)

# Display mean score
display_scores(ridge_rmse_scores)
```

Therefore, after evaluating three different models using cross validation, all the models have performed quite similar. However, as far as score is concerned, **Lasso regression** model **(alpha = 0.1)** has outperformed the other two models with marginal difference. Thus, we will use this model to proceed with the upcoming tasks. Moreover, any one of these models can be used since their difference in performance is insignificant. Perhaps, another model might perform better than the one listed here when used in a different scenario.

## INFERENCE SCRIPT

In order for the model to be applied to any data, we will export both the pipeline and the model that performs the best among those mentioned above. This can be achieved using the **Pickle** module which supports reusability of models and functions and their sharing across various platforms and programmes.

```python
import pickle

# Saving the best model as pickle file
with open('ranFor.pkl', 'wb') as f:
    pickle.dump(randomForest_final, f)

# Saving the pipeline and best model as pickle files
with open('pipeline.pkl', 'wb') as f:
    pickle.dump(pipeline, f)
    
# Load pipeline and best model from pickle files
with open('pipeline.pkl', 'rb') as f:
    processing = pickle.load(f)

with open('ranFor.pkl', 'rb') as f:
    test_model = pickle.load(f)
```

```python
# Function for generating inference script
def inference_script(test_data, val_data, preprocessor = processing, model = test_model):
    
    # Process & Predict
    data_process = preprocessor.fit_transform(test_data)
    production_preds = model.predict(data_process)
    
    # Retreive Predictions
    preds_data = pd.DataFrame(production_preds, columns = ['DeathRate'])
    
    # Saving the predictions
    preds_data.to_csv('predictions_reg.csv', index=False)
    print("The predictions have been saved in the directory!\n")
    
    # Evaluting the model using RMSE
    prod_mse=mean_squared_error(val_data, production_preds)
    prod_rmse=np.sqrt(prod_mse)
    print("Model RMSE: {}" .format(prod_rmse))

    return preds_data.head(10)
```

```python
# Inference script
# Input your 'test' data here
Test = test_data

# Input your 'validation' data here
Validation = test_targets

inference_script(Test, Validation)
```
